```python
import numpy as np
import matplotlib.pyplot as plt
import math


def VALUE(x):
    
    if(x<-1 or x>1):
        y=0
    else:
        y=math.cos(40*x)+math.sin(10*x)
    return y

VECTORIZED_VALUE=np.vectorize(VALUE)
x=np.linspace(-1,2,num=1000)
plt.plot(x,VECTORIZED_VALUE(x))
```

```python
import random

def INITIAL_STATE(problem):
    initial=random.uniform(-1,1)
    return(initial)


def SCHEDULE(T,t):

    if(T>0.0001):
        T=(0.995**t)*T
    else:
        T=0
    
    return(T)

def RANDOM_SUCCESSOR(current):
    
    if(current-0.05<-1):
        w=random.uniform(0,0.1)
    elif(current+0.05>1):
        w=random.uniform(-0.1,0)
    else:
        w=random.uniform(-0.1,0.1)
    val=current+w
    return(val)


def SIMULATED_ANNEALING(problem,schedule):
    
    current=INITIAL_STATE(problem)
    
    t=1
    T=10
    while(t>0):    
        T=SCHEDULE(T,t)
        if(T==0):
            return current
        next_1=RANDOM_SUCCESSOR(current)
        E=VALUE(next_1)-VALUE(current)
        if(E>0):
            current=next_1
        else:
            if(math.exp(E/T)>random.uniform(0,1)):
                current=next_1
        t=t+1
        
x=SIMULATED_ANNEALING("prob",0.995)
y=VALUE(x)
print("The global maximum calculated by Simulated Annealing is at "+str(x)+ " and the value of the maxima is "+str(y))
```
